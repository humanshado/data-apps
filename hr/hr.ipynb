```python
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
import numpy as np
import dash_bootstrap_components as dbc
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df.info()
```

```python
def clean_df(data, col, col_name):
    l = []
    for i in col.items():
        temp = i[1].split(',')
        temp = ''.join(map(str, temp))
        l.append(temp)

    col = pd.to_numeric(pd.Series(l))
    data[col_name] = col
```

```python
col_names = ['Locals', 'Expatriates', 'Headcount', 'Incentive Paid', 'Overtime Paid']

for col_name in col_names:
    col = df[col_name]
    clean_df(df, col, col_name)
    
df.info()
```

```python
cleaned_df = df
cleaned_df
```

```python
mask = cleaned_df.Month == 'Jan'
m_selection = cleaned_df[mask][['Locals', 'Expatriates']]
m_selection
```

```python
m_selection = m_selection.melt(var_name='Staff', value_name='#StaffCount')
m_selection
```

#### Dashboard

```python
app = JupyterDash(__name__, external_stylesheets=[dbc.themes.YETI])

colors = {
    'yaleblue': '#064789',
    'steelblue': '#427aa1',
    'aliceblue': '#ebf2fa',
    'claret': '#840032',
    'gamboge': '#e59500'
    
}

app.layout = dbc.Container([
                html.H2('ABC Human Resources Dashboard', className='mb-5 text-center font-weight-bold'),
                dbc.Row([
                    dbc.Col([
                        html.Label("Select a month"),
                        dcc.Dropdown(
                            id='month-dropdown',
                            options=[{'label': str(i), 'value': str(i)} for i in cleaned_df.copy().Month],
                            value='Jan'
                        ),
                    ], width=1),
                    
                    dbc.Col([
                        dbc.Card([
                            dbc.CardBody([
                                html.H6('Staff', className='card-title text-center'),
                                html.Hr(),
                                dcc.Graph(id='pie-chart', figure={}, config={'displayModeBar': False}, style={'height': 250}),
                           ])
                        ], className='mr-1')
                    ], width={'size': 2, 'offset': 1}),

                    dbc.Col([
                        dbc.Card([
                            dbc.CardBody([])
                        ])
                    ], width=1),

                    dbc.Col([
                        dbc.Card([
                            dbc.CardBody([])
                        ])
                    ], width=1),

                    dbc.Col([
                        dbc.Card([
                            dbc.CardBody([])
                        ])
                    ], width=2),

                    dbc.Col([
                        dbc.Card([
                            dbc.CardBody([])
                        ])
                    ], width=3),

                    dbc.Col([
                        dbc.Card([
                            dbc.CardBody([])
                        ])
                    ], width=1),
                ], no_gutters=True, className='mb-2'), #row1
        ], fluid=True) #container

# Pie Chart
@app.callback(
    Output('pie-chart', 'figure'),
    Input('month-dropdown', 'value'),
)
def update_pie(month_chosen):
    dff = cleaned_df.copy()
    mask = dff.Month == month_chosen
    my_selection = dff[mask][['Locals', 'Expatriates']]
    select_melted = my_selection.melt(var_name='Staff', value_name='#StaffCount')
    fig_pie = px.pie(select_melted, values='#StaffCount', names='Staff', 
                 template='ggplot2', color_discrete_sequence=[colors['yaleblue'], colors['aliceblue']])
   
    fig_pie.update_layout(margin=dict(l=0, r=0, t=0, b=0), legend=dict(orientation="h", yanchor="bottom", y=1.02,
                                                                       xanchor="right", x=1))
    
    return fig_pie


app.run_server(mode='jupyterlab')
```
